```python
#FUNCIONES
def is_prime(n: int):           #Comprueba si un número n es primo.
    if n <= 1:
        return False            #Todo número es divisible por 1 y por 0
    for i in range(2, n, 1):    # Revisar desde 2 hasta n-1
        if n % i == 0:          # Si encuentra un divisor
            return False        # No es primo
    return True                 # Si no encuentra divisores, es primo

def prime_list(limit):          #Genera una lista de números primos hasta un límite limit.  
    lista_primos= []
    if limit <= 1:
        return 
    for num in range(2, limit, 1):
        primo = True
        for i in range(2, num, 1):      
            if num % i == 0:          
                primo = False
                break
        if primo:
            lista_primos.append(num)
    return lista_primos

def check_palindrome(primos: int):          #Busca palíndromos en una lista de números primos (se leen igual de adelante hacia atrás).
    listapalindromos = []
    for num in primos:                      # Iterar sobre cada número primo
        numcadena = str(num)                # Convertir número a cadena
        if numcadena == numcadena[::-1]:    # Verificar si es igual al revés
            listapalindromos.append(num)
    return listapalindromos

def categorize_prime(primos: int):          # Clasifica los números primos en "pequeños" (menores que 10), "medianos" (de 10 a 99) o "grandes" (100 o más).
    listatamano= []
    print("\n4.   Rango de valores\n      -Menores que 10: ", end="")
    for num in primos:
        if num < 10:
            print(num, end= " ")
    print("\n      -Entre 10 y 99: ", end="")
    for num in primos:
        if num >= 10 and num < 99:
            print(num, end= " ")
    print("\n      -Mayores o igual que 100: ", end="")
    for num in primos:
        if num >= 100:
            print(num, end= " ")
                    
#VARIABLES
lista = [1, 7, 8, 11, 23, 56, 57, 911, 10101]
limit = 150
print(f"LISTA: {lista}\n")

#APARTADO 1
print("1.   Números primos que hay en la lista: ", end="")
for numero in lista:
    if is_prime(numero):
        print(numero, end= " ") # Imprime en la misma línea con un espacio
print("\n")    

#APARTADO 2
primos = (prime_list(limit))
print(f"2.   Los primeros {limit} numeros primos que existen son:", primos, "\n")

#APARTADO 3
print(f"3.   Los primos palíndromos son:",check_palindrome(primos))  

#APARTADO 4
categorize_prime(primos)
```
