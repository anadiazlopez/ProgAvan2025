```python
frase1 = input("Primera frase: ").upper()
frase2 = input("Segunda frase: ").upper()
frase3 = input("Tercera frase: ").upper()

lista = [frase1, frase2, frase3]

def imprimir_cadena():
    while True:
        letra = input("Ingrese una letra: ").upper()
        if len(letra) == 1:
            break
        else:
            print("Error: debe ingresar solo un carácter.")
    for frase in lista:
        if frase[0] == letra:
            print(f"{frase}")

def contar_subcadena():
    subcadena = input("Ingrese una subcadena a buscar: ").upper()
    contador = 0
    for frase in lista:
        if subcadena in frase:
            contador += 1
    if contador == 1:
        print(f"La subcadena '{subcadena}' aparece en una frase.\n")
    else:
        print(f"La subcadena '{subcadena}' aparece en {contador} frases.\n")

def buscar_cadena():
    n1 = len(frase1)
    n2 = len(frase2)
    n3 = len(frase3)

    if n1 == n2 == n3:
        print("Las tres cadenas tienen el mismo tamaño.\n")
    elif n1 > n2 and n1 > n3:
        print("La primera cadena es la más larga.")
        if n2 > n3:
            print("La tercera cadena es la más corta.\n")
        elif n2 < n3:
            print("La segunda cadena es la más corta.\n")
        else:  # n2 == n3
            print("La segunda y la tercera cadena son las más cortas.\n")
    elif n2 > n1 and n2 > n3:
        print("La segunda cadena es la más larga.")
        if n1 > n3:
            print("La tercera cadena es la más corta.\n")
        elif n1 < n3:
            print("La primera cadena es la más corta.\n")
        else:  # n1 == n3
            print("La primera y la tercera cadena son las más cortas.\n")
    elif n3 > n1 and n3 > n2:
        print("La tercera cadena es la más larga.")
        if n1 > n2:
            print("La segunda cadena es la más corta.\n")
        elif n1 < n2:
            print("La primera cadena es la más corta.\n")
        else:  # n1 == n2
            print("La primera y la segunda cadena son las más cortas.\n")
    elif n1 == n2 and n1 > n3:
        print("La primera y la segunda cadena son las más largas.")
        print("La tercera cadena es la más corta.\n")
    elif n1 == n3 and n1 > n2:
        print("La primera y la tercera cadena son las más largas.")
        print("La segunda cadena es la más corta.\n")
    elif n2 == n3 and n2 > n1:
        print("La segunda y la tercera cadena son las más largas.")
        print("La primera cadena es la más corta.\n")

def mismo_objeto():
    print("\nLista de frases:")
    for i, frase in enumerate(lista):
        print(f"{i}: {frase}")

    i1 = int(input("Ingrese el índice de la primera cadena: "))
    i2 = int(input("Ingrese el índice de la segunda cadena: "))

    if lista[i1] is lista[i2]:
        print("Son el mismo objeto en memoria (is).")
    else:
        print("No son el mismo objeto en memoria.")
     
def mayor_que_10():
    i = 1
    encontrado = False   

    for frase in lista:
        if len(frase) > 10:
            print(f"La cadena {i} tiene más de 10 caracteres.\n")
            encontrado = True
        i += 1
    else:
        if not encontrado:
            print("Ninguna cadena tiene más de 10 caracteres.\n")

def menu():
    print("\nMENÚ DE OPERACIONES")
    print("1. Imprimir cadena.")
    print("2. Contar subcadenas.")
    print("3. Buscar cadena.")
    print("4. Mismo objeto.")
    print("5. Cadena con mas de 10 caracteres.")
    print("7. Salir")
    print("\n")

def main():
    while True:
        menu()
        opcion = input("Seleccione una opción: ")
        if opcion == "1":
            imprimir_cadena()
        elif opcion == "2":
            contar_subcadena()
        elif opcion == "3":
            buscar_cadena()
        elif opcion == "4":
            mismo_objeto()
        elif opcion == "5":
            mayor_que_10()
        elif opcion == "6":
            print("¡Gracias por usar el programa! Hasta pronto.")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.\n")


main()


```
